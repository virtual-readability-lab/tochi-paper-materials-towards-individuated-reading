```python
# Load data
data_speed_raw = pd.read_csv('Data/reading_speed.csv')
display(data_speed_raw)
```

```python
df_speed = (
    data_speed_raw
    .melt(id_vars = 'id_user', var_name = 'font', value_name = 'wpm')
    .dropna(subset = ['wpm'])
    .sort_values(by = ['id_user', 'font'])
)

display(df_speed)



#df_speed.assign(double = lambda x: x.wpm * 2, half = lambda x: x.double - 2)
#df_speed.assign(double = df_speed.wpm * 2)
```
